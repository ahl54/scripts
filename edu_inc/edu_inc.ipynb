```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sp
# from math import log

xls = pd.ExcelFile('PPCS_incarceration_by_state.xls')
edu = pd.read_excel(xls, 'PPCS', parse_cols='A,B').dropna()
inc = pd.read_excel(xls, 'US_incarceration', parse_cols='B,C').dropna()
```

```python

x = list(edu['Per Pupil Current Spending (PPCS)'])
y = list(inc['Yearend'])
# x = [log(a,2) for a in x]
# y = [log(b,2) for b in y]
labels = list(edu['State'])

slope, intercept, r_value, p_value, std_err = sp.linregress(x, y)
r = r'$r\_sq={}$'.format(r_value*r_value)

fig, ax = plt.subplots(figsize=(20, 10))
ax.scatter(x, y, s=10)
ax.annotate(r, xy=(2, 1), xytext=(3, 1.5))
for i, label in enumerate(labels):
    ax.annotate(label, (x[i],y[i]), fontsize=12)
```

```python
plt.title("2013 U.S. Per Pupil Current Spending (PPCS) vs incarceration by state")
plt.xlabel("Per Pupil Current Spending (PPCS) $USD")
plt.ylabel("Yearend in prison or jail")
plt.show()
```
